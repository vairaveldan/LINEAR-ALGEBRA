```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[1, 6],
              [5, 2]])

## Define the vectors u and v
u = np.array([6, -5])
v = np.array([3, -2])

## Calculate  matrix multiplication of A & u and A & v
Au = np.dot(A, u)
Av = np.dot(A, v)

## Plot the vectors
origin={'x':0,'y':0} # just a dict for origin point
plt.figure(figsize=(10, 10)) # figure size of output visualization :figsize=(x units , y units)

# plt.quiver() function in matplotlib is used to plot the arrows. 
# It takes several parameters to customize the appearance of the arrows.
# ex: plt.quiver(X , Y, U, V, angles='xy' , scale_units='xy', scale=?, color=?, label=?)
#here X, Y: These parameters specify the starting coordinates of the arrow.
#here U, V:These parameters specify the ending coordinates of the arrow.
#here  angles='xy' is commonly used for 2D plots, where the vectors lie in the xy-plane.
#here scale_units='xy' which specifies the type of units to use for the vector scale
#here scale=? which specifies the scaling factor applied to the length of the arrows. An arrow of [0,1] with 1 scaling factor of one equals to an array of [0,2] with scaling factor of two
#here color and label as usual represent color and label of the array

plt.quiver(origin['x'],origin['y'] , u[0], u[1],angles='xy',  scale_units='xy', scale=1, color='b', label='u')
plt.quiver(origin['x'],origin['y'] , v[0], v[1], angles='xy', scale_units='xy', scale=1, color='g', label='v')
plt.quiver(origin['x'],origin['y'] , Au[0], Au[1], angles='xy', scale_units='xy', scale=1, color='r', label='A*u')
plt.quiver(origin['x'],origin['y'] , Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='m', label='A*v')

## Set plot limits and labels

# Set the limits for the x and y axes of the plot
plt.xlim(-25, 10)
plt.ylim(-10, 25)

# Label the x and y axes
plt.xlabel('x')
plt.ylabel('y')

# just a title for the plot 
plt.title('Visualization of eigenvectors')

# grid is a x&y line you see in every 5 units
plt.grid(color = 'gray')

# Show labels with loc-> location where the labels have to place and as usual title for labels
plt.legend(loc='upper right', title='Vectors')

# Show plot
plt.show()
```
